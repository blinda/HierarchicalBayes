```python
import os
import json
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
```

```python
cwd = os.getcwd()

dataPX = open(cwd+'/totSample/pxnorm_distrSNR.json')
dataRB = open(cwd+'/totSample/rdblurnorm_distrSNR.json')

pdPX = json.load(dataPX)
pdRB = json.load(dataRB)

RPX = np.zeros(len(pdPX))
RRB = np.zeros(len(pdRB))

for i, k in enumerate(pdPX.keys()):
    RPX[i] = np.mean(pdPX[k])
    RRB[i] = np.mean(pdRB[k])
```

```python
fig = plt.figure(figsize=(12,12))

plt.hist(RRB, color='crimson', alpha=0.6, label='Broad reflection, posterior mean', bins=20)
plt.hist(RPX, color='dodgerblue', alpha=0.6, label='Narrow reflection, posterior mean', bins=20)

plt.axvline(np.mean(RRB), color='crimson', 
            label='Broad reflection,\n population mean: '+np.str(np.around(np.mean(RRB),2)))
plt.axvline(np.mean(RPX), color='royalblue',
           label='Narrow reflection,\n population mean: '+np.str(np.around(np.mean(RPX),2)))

plt.xlabel('$\mathrm{log_{10}(R)}$',fontsize=24)
plt.ylabel('$\mathrm{Number\; of\; objects}$', size=24)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

legend = plt.legend(loc='best', scatterpoints=1, numpoints=1, prop={'size':16})
legend.set_frame_on(False)

plt.savefig('Rmean.png')
plt.savefig('Rmean.pdf', dpi=200)
```
