```python
import os
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
from Stan_output import read_output, bins
```

```python
L, NH, z = bins('L'), bins('NH'), bins('z')
pxL, errpxL, spxL, errspxL = read_output('pxnorm', 'L')
pxNH, errpxNH, spxNH, errspxNH = read_output('pxnorm', 'NH')
pxz, errpxz, spxz, errspxz = read_output('pxnorm', 'z')

rbL, errrbL, srbL, errsrbL = read_output('rdblurnorm', 'L')
rbNH, errrbNH, srbNH, errsrbNH = read_output('rdblurnorm', 'NH')
rbz, errrbz, srbz, errsrbz = read_output('rdblurnorm', 'z')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(19,6))


ax1.errorbar(L, rbL, yerr= errrbL, marker='s', markersize =3 ,color='C3', 
             alpha=0.7, label=r'Broad reflection, population mean', fmt='')
ax2.errorbar(z, rbz, yerr= errrbz, marker='s', markersize =3 ,color='C3', 
             alpha=0.7, label=r'Broad reflection, population mean', fmt='')
ax3.errorbar(NH, rbNH, yerr= errrbNH, marker='s', markersize =3 ,color='C3', 
             alpha=0.7, label=r'Broad reflection, population mean', fmt='')

ax1.errorbar(L, pxL, yerr= errpxL, marker='v', linestyle='--', markersize =3 ,color='C0',
             alpha=0.7, label=r'Narrow reflection, population mean', fmt='')
ax2.errorbar(z, pxz, yerr= errpxz, marker='v', linestyle='--', markersize =3 ,color='C0', 
             alpha=0.7, label=r'Narrow reflection, population mean', fmt='')
ax3.errorbar(NH, pxNH, yerr= errpxNH, marker='v', linestyle='--', markersize =3 ,color='C0', 
             alpha=0.7, label=r'Narrow reflection, population mean', fmt='')


ax1.set_xlabel('$\mathrm{log_{10}(L_{2-10})\, erg/s}$', fontsize=18)
ax2.set_xlabel('$\mathrm{Redshift}$',fontsize=18)
ax3.set_xlabel('$\mathrm{log_{10}(N_H)\, [atoms\, cm^{-3}]}$',fontsize=18)

ax1.set_ylabel('$\mathrm{log_{10}(Reflection\, fraction)}$', size=18)
ax1.tick_params(labelsize=16)
ax2.tick_params(labelsize=16)
ax3.tick_params(labelsize=16)

legend = plt.legend(loc='best', scatterpoints=1, numpoints=1, prop={'size':14})
legend.set_frame_on(False)

plt.savefig('mean_RLzNHkn7.png')
plt.savefig('mean_RLzNHkn7.eps', rasterized=True)
```
